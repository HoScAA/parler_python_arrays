```python
import numpy as np
import matplotlib.pyplot as plt
n = 5
N = 2**n
a = np.zeros(N)
a[0] = 1
a[8] = 1
a[16] = 1
a[24] = 1
a = a*1/np.sqrt(4)
a
```

```python
c = np.fft.fft(a)
plt.plot(c)
```

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.circuit.library import QFTGate
from qiskit.visualization import array_to_latex, plot_bloch_multivector

psi_0 = Statevector(a)
qc = QuantumCircuit(n)
qc.initialize(psi_0)
qc = qc.compose(QFTGate(n))
qc.draw(output='mpl')
psi_1 = Statevector(qc)
```

```python
plt.plot(psi_1.data)
```
